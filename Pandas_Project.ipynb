**# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017**

```python
print(df.query('year == 2017')['Sales'].agg(['sum', 'mean', 'std']))
```

**# TODO 06 - which Segment has the highest profit in 2018**

---



```python
profit_by_segment = df.query('year == 2018')\
                .groupby('Segment')['Profit']\
                .sum().reset_index()\
                .sort_values('Profit', ascending=False)
print(profit_by_segment)
```

**# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019**

```python
df_range = df[(df['Order Date'] >= "2019-04-15") & (df['Order Date'] <= "2019-12-31")]
df_range.groupby('State')['Sales'].sum().sort_values().head().reset_index()
```

**# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% **



```python
df_2019 = df[(df['Order Date'] >= "2019-01-01") & (df['Order Date'] <= "2019-12-31")]
sales_WC = df_2019.query('Region == "West" | Region == "Central"')['Sales'].sum()
total_sales = df_2019['Sales'].sum()
print(sales_WC)
print(total_sales)
print("%.2f%%" % (100*(sales_WC/total_sales)))
```

**# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020**

```python
year19_20 = df[(df['Order Date'] >= "2019-01-01") & (df['Order Date'] <= "2020-12-31")]
Total_quantity = year19_20.groupby('Product Name')['Quantity'].sum().reset_index()
Total_sales = year19_20.groupby('Product Name')['Sales'].sum().reset_index()
quantity_sales = pd.merge(Total_quantity, Total_sales, on="Product Name")
```

```python
quantity_sales.sort_values('Quantity',ascending=False).head(10)
```
